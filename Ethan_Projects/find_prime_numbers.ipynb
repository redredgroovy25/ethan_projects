```python
import time
import math
import pandas
```

```python
def calc_primes(first_num, second_num):
    first_num = int(first_num)
    second_num = int(second_num)
    print('Finding prime numbers between ',first_num,' and ', second_num,' ....')
    list_primes = []
    test_number = first_num
    
    for test_number in range(first_num, second_num+1):
        if test_number != 2 and test_number % 2 == 0:
            continue
        else:
            range_top = (test_number//2)+1  # half the test number (ish)
            factor_count = 0
            for j in range(1, range_top): # don't test any value that is larger than half your test_number
                if test_number % j == 0:
                    factor_count += 1
                    j += 1
            if factor_count == 1:
                list_primes.append(test_number)
                
    print('... All done!!')
    return list_primes
```

```python
print('Let''s find some prime numbers.')

print('Enter the lower bound of the range that you would like to check.')
lower_bound = input()
print('Now enter the upper bound of the range you would like to check.')
upper_bound = input()

start_time=time.time()
list_primes = calc_primes(lower_bound, upper_bound)

print(time.time()-start_time,' seconds to complete calculation.')
    
print('The largest prime number between ',lower_bound,' and ',upper_bound,' is:')
print(list_primes[len(list_primes)-1])

save_it = input('Would you like to save all the numbers to a file (Y or N)?')

if save_it=='Y' or save_it=='y':
    df = pandas.DataFrame(data= list_primes)
    df.to_csv("./primes.csv", sep=',',index=False)
    print('The file is called primes.csv.')
else:
    print('Okay, no file for you!')


    
```
