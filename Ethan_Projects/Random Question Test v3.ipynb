## Random Question Test

Here is a little code that lets you create your own quiz questions in an excel spreadsheet.
You can also assign each question a number of points and provide as many multiple choice answers as you would like.

When you run it, the code will randomize the order of the questions and the order of the answer choices.  It will also allow you to select the answer by multiple choice (providing the number of the answer).

At the end of the quiz, you will be able to see your score out of the possible total maximum points.

Enjoy!

```python
# Import a few helpful modules

from numpy import random   # for randomizing questions and answers
import pandas as pd   # for importing my excel list of questions
import ast   # this is used for converting a string into a python list. it's pretty cool, i just learnded about it.
```

```python
#I'm going to keep score for this quiz, so i'll create  a score counter.

total_score = 0
```

```python
class Question:
    """This is my Question Class.   
    It contains a prompt with the actual question and a list of tuples for answers.
    Each answer contains along with it whether it is the right answer or not.
    Each question also has its own points value.  We will keep score and produce a final point total.
    ask_question() - prints the question for the quiz taker
    list_answers() - provides the list of possible answers
    check_answer() - checks the user's answer and scores it"""
    
    
    prompt = ""
    answers = list(tuple())
    question_value = 0
    
    def __init__(self,prompt,answers,question_value):
        self.prompt=prompt
        self.answers = answers
        self.question_value = question_value
        
    def ask_question(self):
        print(self.prompt)
        
    def list_answers(self):
        random.shuffle(self.answers)
        print('Please type the choice number corresponding to the correct selection: \n')
        i=0
        while i < len(self.answers):
            new_answer =list(self.answers[i])
            new_answer.append(str(i+1))
            print(new_answer[2],new_answer[0])
            self.answers[i] = new_answer
            i+=1
        print()
            
    def check_answer(self,user_answer):
        for answer in self.answers:
            if answer[2]==user_answer and answer[1]==True:
                print('\n You got it! \n')
                return int(self.question_value)
        print('\n Wrong answer. \n')
        return 0
        
       
                
```

```python
# Here, I import the list of quiz questions I would like to ask from an excel file. 
# If you would like to add to the file, you can just open it and add new questions.    
# However, the answers have to be in the correct format or you will get an error when you try to use them.

quiz_questions = pd.read_excel(io = 'Quiz_Questions.xlsx')

# Un-comment this and run it if you want to see the questions ahead of time, but that is cheating.
#quiz_questions
```

```python
# I am creating the list of questions based on the contents of the excel spreadsheet I have imported above.
# Pretty neat-o.

quiz_question_list=[]
i=0

while i < len(quiz_questions):
    prompt = quiz_questions['prompt'][i]
    answers_string = quiz_questions['answer_list'][i]
    answers = ast.literal_eval(answers_string)
    question_value = quiz_questions['points'][i]
    quiz_question_list.append(Question(prompt, answers, question_value))
    i+=1
    
```

```python
max_points = 0
for question in quiz_question_list:
    max_points+=question.question_value
```

###  And Now....  Time to start the Quiz!!!

First, I provide a little intro, then we launch into the questions, which are pretty great, even if I do say so myself....

```python
#after I create the list of questions, I can randomly shuffle them so they will be asked in a random order.
random.shuffle(quiz_question_list)

print(f'Welcome to my quiz.  I will ask you {len(quiz_question_list)} questions.')
print(f'The total value of points you can win is {max_points}.')
print('Lets go! \n\n')
```

```python
for question in quiz_question_list:
    question.ask_question()
    question.list_answers()
    user_answer = input()
    points = question.check_answer(user_answer)
    total_score += points


print('total score = ',total_score, 'out of ',max_points,' possible points.')
    
```
