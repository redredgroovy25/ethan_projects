```python
# First, i need to import random so i can shuffle my questions and answers.

from numpy import random
import pandas as pd
import json 
```

```python
#I'm going to keep score for this quiz, so i'll create  a score counter.
total_score = 0
```

```python
class Question:
    """This is my Question Class.   
    It contains a prompt with the actual question and a list of tuples for answers.
    Each answer contains along with it whether it is the right answer or not.
    Each question also has its own points value.  We will keep score and produce a final point total.
    ask_question() - prints the question for the quiz taker
    list_answers() - provides the list of possible answers
    check_answer() - checks the user's answer and scores it"""
    
    
    prompt = ""
    answers = list(tuple())
    question_value = 0
    
    def __init__(self,prompt,answers,question_value):
        self.prompt=prompt
        self.answers = answers
        self.question_value = question_value
        
    def ask_question(self):
        print(self.prompt)
        
    def list_answers(self):
        random.shuffle(self.answers)
        print('Please type the choice number corresponding to the correct selection: \n')
        i=0
        while i < len(self.answers):
            new_answer =list(self.answers[i])
            new_answer.append(str(i+1))
            print(new_answer[2],new_answer[0])
            self.answers[i] = new_answer
            i+=1
        print()
            
    def check_answer(self,user_answer):
        for answer in self.answers:
            if answer[2]==user_answer and answer[1]==True:
                print('\n You got it! \n')
                return int(self.question_value)
        print('\n Wrong answer. \n')
        return 0
        
       
                
```

```python
# This is my list of questions .    In a future version, 
# let's figure out how to specify these in a file so that we can add them more easily!

quiz_question_list = []

prompt1 = 'What is my name?'
answers1 = [('Mummy',True),('Silly Billy',False),('Ignatius P. Seward III', False)]
question_value1 = 20
quiz_question_list.append(Question(prompt1, answers1, question_value1))

prompt2 = 'How old am I?'
answers2 = [('511',False),('51',True),('12',False)]
question_value2 = 30
quiz_question_list.append(Question(prompt2, answers2, question_value2))

prompt3 = 'What is the airspeed velocity of an unladen swallow'
answers3 = [('African or European swallow',True),('Streuth!',False),('IDK',False)]
question_value3 = 1
quiz_question_list.append(Question(prompt3,answers3,question_value3))
```

```python
max_points = 0
for question in quiz_question_list:
    max_points+=question.question_value
```

```python
#after I create the list of questions, I can randomly shuffle them so they will be asked in a random order.
random.shuffle(quiz_question_list)

print(f'Welcome to my quiz.  I will ask you {len(quiz_question_list)} questions.')
print(f'The total value of points you can win is {max_points}.')
print('Lets go! \n\n')
```

```python
for question in quiz_question_list:
    question.ask_question()
    question.list_answers()
    user_answer = input()
    points = question.check_answer(user_answer)
    total_score += points


print('total score = ',total_score, 'out of ',max_points,' possible points.')
    
```
